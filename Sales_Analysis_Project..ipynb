```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.columns = df.columns.str.strip()

df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True, errors='coerce')
```

```python
df[['Order Date', 'Ship Date']].head()
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(category_sales)
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Products by Sales")
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales by Region")
plt.show()
```

```python
df['Month'] = df['Order Date'].dt.month

monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.show()
```

```python
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)

city_sales.plot(kind='bar')
plt.title("Top Cities by Sales")
plt.show()
```

```python
-Technology category generates highest revenue
-Certain products contribute significantly to total sales
-West region has higher sales compared to others
-Sales show variation across months indicating seasonal trends
-Major cities contribute a large portion of total revenue
```
